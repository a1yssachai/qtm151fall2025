### c) For-loops + Plotting

   - Create an array of x-values with 100 evenly spaced points between $0$ and $2\pi$ <br>using `np.linspace()`. **Hint**: type ``help(np.linspace)`` in a separate cell to see how it is used

   - Create a vector named ```vec_y1``` where each element is computed from the formula <br> $ y_k = \sin(x_k)$. (Here $x_k$ is the $k$-th element of the array you created above, and $y_k$ is the $k$-th element of vec_y1)

   - Create a vector ```vec_y2``` where each element is computed from the formula <br> $ y_k = \cos(x_k)$.

   - Create a list ```list_vectors``` with elements ```vec_y1``` and ```vec_y2```. (So a list with two components, both being numpy arrays)

   - Create a for-loop that goes through each element of ```list_vectors``` and plots each <br>list of $y$ values on the vertical axis against $x$ on the horizontal axis.

Note: The final output should be a 2-dimensional plot with the values from vec_y1 and vec_y2 on the vertical axis, and the $x$ values on the horizontal axis. **Remember to label the axes and add a title to the plot**.

````python
# Write your answer here:
import matplotlib.pyplot as plt

x_vals = np.linspace(start = 0, stop = 2*np.pi, num = 100)

vec_y1 = np.sin(x_vals)
vec_y2 = np.cos(x_vals)

list_vectors = [vec_y1, vec_y2]

for vec in list_vectors:
    plt.plot(x_vals, vec)

plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Graph of Sine and Cosine Functions for $0 \leq x \leq 2\pi$")
plt.show()

# extra stuff
np.random.seed(1234)   # keep results reproducible
n = 1000               # number of draws from each distribution

# Generate random samples
normal_samples  = np.random.normal(loc=0, scale=1, size=n)
uniform_samples = np.random.uniform(low=-3, high=3, size=n)

# Create 1 row and 2 columns of subplots
fig, subfig_array = plt.subplots(nrows=1, ncols=2, figsize=(8,3))

# Left histogram: normal
subfig_array[0].hist(normal_samples, bins=30, color="lightblue")
subfig_array[0].set_title("Normal Distribution (mean=0, sd=1)")
subfig_array[0].set_xlabel("Sample Value")
subfig_array[0].set_ylabel("Frequency")

# Right histogram: uniform
subfig_array[1].hist(uniform_samples, bins=30, color="salmon")
subfig_array[1].set_title("Uniform Distribution (-3, 3)")
subfig_array[1].set_xlabel("Sample Value")
subfig_array[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
````

````python
# Look over multiple CSV files and store results

file_list = [
    "practice_quiz_data/china_co2_emissions.csv",
    "practice_quiz_data/germany_co2_emissions.csv",
    "practice_quiz_data/india_co2_emissions.csv",
    "practice_quiz_data/indonesia_co2_emissions.csv",
    "practice_quiz_data/iran_islamic_rep_co2_emissions.csv",
    "practice_quiz_data/japan_co2_emissions.csv",
    "practice_quiz_data/korea_rep_co2_emissions.csv",
    "practice_quiz_data/russian_federation_co2_emissions.csv",
    "practice_quiz_data/saudi_arabia_co2_emissions.csv",
    "practice_quiz_data/united_states_co2_emissions.csv"
]

# Step 2: Create an empty list to store the mean emissions
mean_emissions = []

# Step 3: Loop over each file
for filename in file_list:
    us_emissions = pd.read_csv(filename)
    mean_val = us_emissions["total_emissions"].mean()
    mean_emissions.append(mean_val)

# for cleam print out!!!
countries = [
    "China", "Germany", "India", "Indonesia", "Iran", "Japan",
    "South Korea", "Russia", "Saudi Arabia", "United States"
]

for country, filename in zip(countries, file_list):
    us_emissions = pd.read_csv(filename)
    mean_val = us_emissions["total_emissions"].mean()
    print(country + " mean emissions:", round(float(mean_val), 2))

# Step 4: Print the list of mean emission values
print(mean_emissions)
````
